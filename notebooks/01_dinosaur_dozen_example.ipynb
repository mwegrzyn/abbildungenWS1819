#  First Visualization Example

### import modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### load dataset

```python
df = pd.read_csv('../data/dDozen.tsv',sep='\t')
```

```python
df.tail()
```

###  show descriptive statistics

```python
df.groupby('dataset').describe()
```

### barplot

```python
plt.figure(figsize=(16,6))
sns.barplot(data=df,x='dataset',y='x',ci=95,palette='rainbow')
sns.despine()
```

### boxplot

```python
plt.figure(figsize=(16,6))
sns.boxplot(data=df,x='dataset',y='x')
sns.despine()
plt.show()
```

### violin plot

```python
plt.figure(figsize=(16,6))
sns.violinplot(data=df,x='dataset',y='x')
sns.despine()
plt.show()
```

### swarmplot

```python
plt.figure(figsize=(16,6))
sns.swarmplot(data=df,x='dataset',y='x')
sns.despine()
plt.show()
```

### regression line

```python
sns.lmplot(data=df,x='x',y='y',col='dataset',scatter=False,fit_reg=True,col_wrap=4)
plt.xlim(0,100)
plt.ylim(0,100)
```

### scatterplot

```python
sns.lmplot(data=df,x='x',y='y',col='dataset',scatter=True,fit_reg=True,col_wrap=4)
plt.xlim(0,100)
plt.ylim(0,100)
```
