# Annotated Scatterplot

### import modules

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

### set defaults for plotting

```python
sns.set_context('talk')
sns.set_style('ticks')
```

### define data

```python
d = {
    'China': {
        'population': 1394880000,
        'co2': 10641789
    },
    'India': {
        'population': 1338860000,
        'co2': 2454968
    },
    'United States': {
        'population': 328082000,
        'co2': 5172336
    }
}
```

```python
countries_df = pd.DataFrame(d).T

# Divide population by billion
countries_df.loc[:, ['population']] = countries_df.loc[:, ['population']] / 1000000000.0
# Divide co2 by million
countries_df.loc[:, ['co2']] = countries_df.loc[:, ['co2']] / 1000000.
```

```python
countries_df
```

### make plot

```python
# Controls size and aspect ratio of the figure
plt.figure(figsize=(8, 8))

# Move through row names of the table
for i in countries_df.index:

    # For each iteration of the loop:

    # Get population value for country i
    x_val = countries_df.loc[i, 'population']
    # Get co2 value for country i
    y_val = countries_df.loc[i, 'co2']

    # Plot the data points at x and y
    plt.plot(x_val, y_val, 'o')
    # Annotate the points
    plt.annotate(
        xy=(x_val, y_val),  # Position of annotation
        s=i  # String printed for annotation
    )

# Range of axes
plt.xlim(0, 1.5)
plt.ylim(0, 15)

# Label the axes
plt.xlabel('population (billion)')
plt.ylabel('CO2 emissions (millon kt)')
# Title of plot
plt.title('carbon dioxide emissions')

sns.despine(
    offset=10,  # Gap between x and y axis at origin
    trim=True  # End axis at last 'tick'
)

# Show the figure
plt.show()
```

### looping:

```python
# Counter to keep track of the progress of the loop

count = 0  # Initalize at zero

# Move through row names of the table
for i in countries_df.index:

    print('\n*********')

    print('count:', count)

    print('index i:', i)

    # For each iteration of the loop:

    # Get population value for country i
    x_val = countries_df.loc[i, 'population']
    # Get co2 value for country i
    y_val = countries_df.loc[i, 'co2']

    print('value at x:', x_val)
    print('value at y:', y_val)

    # Increase the counter by one
    count += 1
```

### version where we only use the loop for annotation

```python
# Controls size and aspect ratio of the figure
plt.figure(figsize=(8, 8))

# Get population value for country i
x_vals = countries_df.loc[:, 'population']
# Get co2 value for country i
y_vals = countries_df.loc[:, 'co2']

# Plot the data points at x and y
plt.plot(x_vals, y_vals, 'o')

# Move through row names of the table
for i in countries_df.index:

    # For each iteration of the loop:

    # Get population value for country i
    x_val = x_vals[i]
    # Get co2 value for country i
    y_val = y_vals[i]

    # Annotate the points
    plt.annotate(
        xy=(x_val, y_val),  # position of annotation
        s=i  # string printed for annotation
    )

# Range of axes
plt.xlim(0, 1.5)
plt.ylim(0, 15)

# Label the axes
plt.xlabel('population (billion)')
plt.ylabel('CO2 emissions (millon kt)')
# Title of plot
plt.title('carbon dioxide emissions')

sns.despine(
    offset=10,  # gap between x and y axis at origin
    trim=True  # end axis at last 'tick'
)

# Show the figure
plt.show()
```

### version where we make a new plot for each iteration of the loop

```python
# Move through row names of the table
for i in countries_df.index:

    # For each iteration of the loop:

    # Make a new figure
    plt.figure(figsize=(8, 8))

    # Get population value for country i
    x_val = countries_df.loc[i, 'population']
    # Get co2 value for country i
    y_val = countries_df.loc[i, 'co2']

    # Plot the data points at x and y
    plt.plot(x_val, y_val, 'o')
    # Annotate the points
    plt.annotate(
        xy=(x_val, y_val),  # position of annotation
        s=i  # string printed for annotation
    )

    # Range of axes
    plt.xlim(0, 1.5)
    plt.ylim(0, 15)

    # Label the axes
    plt.xlabel('population (billion)')
    plt.ylabel('CO2 emissions (millon kt)')
    # Title of plot
    plt.title('carbon dioxide emissions')

    sns.despine(
        offset=10,  # gap between x and y axis at origin
        trim=True  # end axis at last 'tick'
    )

    # Show the figure
    plt.show()
```
