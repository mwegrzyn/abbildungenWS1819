```python
sns.scatterplot(data=age_df, x='ageBin', y='ageFeel/age')
```

```python
ratio_df = pd.DataFrame(age_df.loc[:, 'ageFeel'] / age_df.loc[:, 'age'])
ratio_df.columns = ['ageFeel/age']
ratio_df.sort_values(by='ageFeel/age', inplace=True, ascending=False)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 8))
sns.barplot(y="ageFeel/age", x=ratio_df.index, data=ratio_df, ax=ax, color="b")
plt.xticks(rotation=90)
plt.show()
```

```python
age_df["ageFeel/age"]=age_df["ageFeel"]/age_df["age"]
age_df
```
