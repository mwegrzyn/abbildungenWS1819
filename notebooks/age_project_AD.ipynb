#  create a new gender variable

```python
#get all different names in gender
df['gender'].unique()
```

```python
# read the first letter of gender and write male or female according to the first letter
x = ()
gender_new = []
for i in [x[0] for x in df ['gender']]:
    if (i =="F"):
        x = 'female'
    elif (i =="f") :
        x = 'female'
    else:
        x = 'male'
    gender_new.append (x)
```

```python
df['gender_new'] = gender_new
```

```python
df.rename(columns={'gender': 'gender_old', 'gender_new': 'gender'}, inplace=True)
```

```python
# add children no children
# read the first letter of gender and write male or female according to the first letter
x = ()
children = []
for i in df['numChild']:
    if (i == 0):
        x = 'no'
    else:
        x = 'yes'
    children.append(x)
df['children']=children
```

```python
sns.scatterplot(data=df, x='age', y='oldBegin')
```

```python
fig, ax = plt.subplots()
#plt.figure(figsize=(16, 16))
fig = (16,16)
sns.regplot(x='age', y='oldBegin', data=df, label = 'old begin')
#ax2 = ax.twinx()
sns.regplot(x='age', y='oldReally', data=df, color='r',label = 'really old')
# Range of axes
plt.xlim(0, 120)
plt.ylim(0, 120)
plt.ylabel('people are old [age]')
plt.xlabel('age participant')
plt.legend()
plt.show()
#sns.scatterplot(data=df, x='age', y='oldBegin')
```

```python
#fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
#fig, ax = plt.subplots()

sns.lmplot('age', # Horizontal axis
           'oldBegin', # Vertical axis
           data=df, # Data source
           fit_reg=True, # Don't fix a regression line
           hue="gender",
           # Set color
           )
# Set x-axis label
plt.xlabel('age')

# Set y-axis label
plt.ylabel('old begin')

sns.lmplot('age', # Horizontal axis
           'oldReally', # Vertical axis
           data=df, # Data source
           fit_reg=True, # Don't fix a regression line
           hue="gender",# Set color
           )

# Set x-axis label
plt.xlabel('age')

# Set y-axis label
plt.ylabel('old really')
```
