## Visualisierung der Daten getrennt für Männer und Frauen

# 1. Frauen

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 8))
sns.barplot(y='ageFeel', x='age', data=female_clean_old, ax=ax, color="b")
plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 8))
sns.barplot(y='ageFeel', x='age', data=female_clean_old, ax=ax, color="b")
sns.swarmplot(data=female_clean_old,x='age',y='ageFeel',color='r')
plt.xticks(rotation=90)
plt.show()
```
